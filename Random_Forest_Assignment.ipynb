**1. Explain Bagging and Boosting methods. How is it different from each other.**

Bagging (Bootstrap Aggregating) is an ensemble learning method that improves the performance of machine learning models by reducing variance. It works by creating multiple subsets of the training data using bootstrapping (random sampling with replacement). Each subset is used to train a separate model independently and in parallel. The predictions from all models are then aggregated, either by majority voting (for classification) or averaging (for regression). Bagging is particularly effective in preventing overfitting, especially for high-variance models like decision trees. Random Forest is a well-known example of Bagging, where multiple decision trees are trained on different subsets of the data and their predictions are combined. One of the key strengths of Bagging is its ability to handle noisy datasets while maintaining computational efficiency through parallel processing. However, Bagging does not significantly reduce bias, which means it may underperform if the base models are too simple.

Boosting, on the other hand, is an ensemble technique designed to reduce both bias and variance by focusing on hard-to-predict samples. Unlike Bagging, Boosting trains models sequentially, where each model corrects the errors made by its predecessor. During this process, misclassified samples are assigned higher weights, forcing subsequent models to prioritize these harder cases. The predictions from all models are combined using weighted voting or averaging. Popular Boosting algorithms include AdaBoost, Gradient Boosting, and XGBoost. Boosting is highly effective at improving model accuracy, especially on imbalanced or complex datasets, as it adapts to the data's underlying patterns. However, it is more prone to overfitting, particularly if the base models are too complex or the algorithm is overtrained. Additionally, Boosting is computationally more expensive than Bagging due to its sequential training process.

In summary, Bagging and Boosting are both ensemble methods but differ significantly in their approach and purpose. Bagging trains models independently to reduce variance and prevent overfitting, making it ideal for high-variance models. Boosting trains models sequentially to reduce both bias and variance, focusing on correcting errors to achieve higher accuracy. While Bagging is less prone to overfitting and computationally efficient, Boosting can achieve better performance but requires careful tuning to avoid overfitting and high computational costs.

**Data** **Visualization**

```python
#step4 : visualizing data using different plots
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Histograms for numerical columns
for column in num_columns:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```

Discount Rate (%) is negatively skewed as it is < 0.

```python
#Boxplots for numerical variables to identify outliers and the interquartile range
for column in num_columns:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x=df[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.ylabel('Values')
    plt.show()
```

```python
#All numericals columns in one boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df[num_columns])
plt.title('Boxplot of Numerical Columns')
plt.xlabel('Numerical Columns')
plt.ylabel('Values')
plt.xticks(rotation=35)
plt.show()
```

For Volume and Discount Rate (%) we dont have outliers. Discount Amount and avg price data points are less scattered, we have less outliers. Total Sales Value and Net Sales Value data points are scattered far way, high value of outliers.

```python
#Bar Chart Analysis for Categorical Column
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
for column in categorical_columns:
    plt.figure(figsize=(8, 6))
    value_counts = df[column].value_counts()
    plt.bar(value_counts.index, value_counts.values, color='skyblue')
    plt.title(f'Bar Chart of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=40)
    plt.show()
```

**Standardization of Numerical** **Variables**

Standardization transforms numerical variables to have a mean of 0 and a standard deviation of 1.
It makes variables comparable by removing their units and scaling them to a similar range.
Formula : z = (x-mu)/sigma where z : standardized value is also called the Z-score,
x : original value,
mu : mean of the column,
sigma : standard deviation of the column

```python
#step5 : import necessary library for standardization
from sklearn.preprocessing import StandardScaler
```

```python
#Histogram plot before standardization
plt.figure(figsize=(8, 6))
plt.title('Before Standardization')
for column in num_columns:
    sns.histplot(df[column], kde=True, label=column)
plt.legend()
plt.show()
```

```python
#standardization
scaler = StandardScaler()
df[num_columns] = scaler.fit_transform(df[num_columns])
```

```python
#histogram plot after standardization
plt.figure(figsize=(8, 6))
plt.title('After Standardization')
for column in num_columns:
    sns.histplot(df[column], kde=True, label=column)
plt.legend()
plt.show()
```

**Conversion of Categorical Data into Dummy** **Variables**

Converting categorical data into dummy variables, also known as one-hot encoding, is necessary because many machine learning algorithms cannot handle categorical data directly.One-hot encoding is a method that turns each category in a column into separate binary columns (0s and 1s). Each new column represents one category. If an observation belongs to a particular category, the corresponding binary column is marked as 1, and all other binary columns for that categorical variable are marked as 0.

```python
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
#step6 : Apply one hot encoding
df_new = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
```

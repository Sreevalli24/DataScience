```python
#fillna is used to replace null values
#handling null values
df.fillna(2,inplace=True)
```

```python
#Checking if numeric values are stored as strings
print(df['LB'].apply(type).value_counts())
print(df['AC'].apply(type).value_counts())
print(df['DS'].apply(type).value_counts())
print(df['DP'].apply(type).value_counts())
print(df['MLTV'].apply(type).value_counts())
print(df['Width'].apply(type).value_counts())
print(df['Tendency'].apply(type).value_counts())
print(df['NSP'].apply(type).value_counts())
```

```python
#Replacing non-numeric entries with NaN
df['LB'] = pd.to_numeric(df['LB'], errors='coerce')
df['AC'] = pd.to_numeric(df['AC'], errors='coerce')
df['DS'] = pd.to_numeric(df['DS'], errors='coerce')
df['DP'] = pd.to_numeric(df['DP'], errors='coerce')
df['MLTV'] = pd.to_numeric(df['MLTV'], errors='coerce')
df['Width'] = pd.to_numeric(df['Width'], errors='coerce')
df['Tendency'] = pd.to_numeric(df['Tendency'], errors='coerce')
df['NSP'] = pd.to_numeric(df['NSP'], errors='coerce')
```

```python
#converting to float
df['LB'] = df['LB'].astype(float)
df['AC'] = df['AC'].astype(float)
df['DS'] = df['DS'].astype(float)
df['DP'] = df['DP'].astype(float)
df['MLTV'] = df['MLTV'].astype(float)
df['Width'] = df['Width'].astype(float)
df['Tendency'] = df['Tendency'].astype(float)
df['NSP'] = df['NSP'].astype(float)
```

```python
#piechart
import matplotlib.pyplot as plt

# Replacing any value that is not 1, 2, or 3 with NaN
df['NSP'] = df['NSP'].apply(lambda x: x if x in [1, 2, 3] else None)

# Drop rows with NaN values in NSP column
df_new = df.dropna(subset=['NSP'])

#count of the valid NSP categories (1, 2, 3)
nsp_counts = df_new['NSP'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(7, 7))  # Set the figure size
plt.pie(nsp_counts, labels=nsp_counts.index, autopct="%2.1f%%") #autopct in python is a string or function used to generate labels inside the wedges 
#showing their numerical values
plt.title('Proportion of Categories in NSP')
plt.show()
```

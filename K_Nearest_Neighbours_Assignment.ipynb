```python
#dropping the 'Animal name' column as it is not useful for prediction
df.drop('animal name', axis=1, inplace=True)
```

```python
#Encoding the target variable 'Type' if it is categorical
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['type'] = label_encoder.fit_transform(df['type'])
```

```python
#splitting the dataset into features and target data
X = df.drop('type',axis=1)
Y = df['type']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
#model building and evaluation
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn
```

```python
knn.fit(X_train,Y_train)
```

```python
y_pred = knn.predict(X_test)
y_pred
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(Y_test,y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

#visualization
# Plot confusion matrix
plt.figure(figsize=(10, 7))
sns.heatmap(confusion_matrix(Y_test,y_pred), annot=True)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
report_df = pd.DataFrame(classification_report(Y_test, y_pred, output_dict=True)).transpose()
plt.figure(figsize=(10,7))
sns.heatmap(report_df.iloc[:-1, :-1], annot=True, cmap='pink')
plt.title('Classification Report')
plt.show()
```

```python
#plotting accuracy
from sklearn.metrics import accuracy_score
plt.figure(figsize=(6,4))
plt.bar(['Accuracy'],accuracy_score(Y_test, y_pred))
plt.ylim(0,1)
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.show()
```

**Interview Questions:**

**1. What are the key hyperparameters in KNN?**

The key hyperparameters in the K-Nearest Neighbors (KNN) algorithm are:

Number of Neighbors (k): This is the number of nearest data points (neighbors) the algorithm considers when making predictions. A small k can make the model sensitive to noise (overfitting), while a large k makes it more generalized but may miss finer details.

Distance Metric (metric): Defines how the distance between points is calculated. Common metrics are:

Euclidean Distance: Straight-line distance (most common).
Manhattan Distance: Sum of absolute differences.
Minkowski Distance: Generalized version of Euclidean and Manhattan.
Weighting Scheme (weights): Determines how neighbors influence predictions:

Uniform: All neighbors have equal influence.
Distance-based: Closer neighbors have more influence than farther ones.
Algorithm (algorithm): Controls how the neighbors are searched:

Brute Force: Checks all points (slower for large datasets).
KD-Tree or Ball-Tree: Faster for large datasets.

**2. What distance metrics can be used in KNN?**

In K-Nearest Neighbors (KNN), different distance metrics are used to measure how close data points are to each other. Here are the common ones in simple terms:

Euclidean Distance: This is the straight-line distance between two points, like measuring with a ruler.

Manhattan Distance: This measures the distance by moving along the grid (up, down, left, or right), like walking in a city with blocks.

Minkowski Distance: A general version of both Euclidean and Manhattan distances, where you can choose how the distance is calculated.

Chebyshev Distance: Looks at the largest difference in any one direction between two points.

Hamming Distance: Counts how many positions are different between two data points. It's great for data like binary (0s and 1s) or categories.

Cosine Distance: Measures how similar two points are based on the angle between them, often used for text or sparse data.

These metrics help KNN decide which points are "closer" and should be considered neighbors. The choice of metric depends on your data type and problem.
